### Ansible Tower

Why ansible tower?  

We cannot have Ansible infrastructure(Control node, credentials, playbooks and vaults) shared across at the enterprise level because.  

* delegating credentials or granting privileged access to all the users poses a security risk
* there is no audit trail of who is doing what
* People from different systems background cannot use it(Suitable for Nix based users) 


### Ansible Tower Architecture

Ansible Tower Web Interface and API  
Users interface with Ansible infrastructure via Ansible Towers web interface or RESTful API  

* Web interface is graphical interface wrapper that executes calls against API  

* Actions available through web interface also available via API  

* RESTful API essential to integrate Ansible with existing software tools and processes  

* Ansible Tower stores data in PostgreSQL back-end database and uses RabbitMQ messaging system  

### Possible tower architectures

* Single machine with integrated database
* Single machine with remote database
* Multi-machine cluster with remote database

### Installing Ansible Tower

Requires RHEL or Centos 7.4 or later with 64 bit architecture.   
Can also be installed on Ubuntu 16.04 LTS (Future builds are not supported)  
4 GB of RAM is a minimum system requirement. With 4 GB 100 nodes can be managed concurrently(A limitation from ansible forks).  
20 GB of disk storage and 10 GB allocation to /var.  

Ansible tower installation itself will be done through ansible playbook.

### Credentials

* Machine
* Network
* scm
* inventory credentials

Private credentials can be created by any individual
Organizational credentials can be created by Org admins
Team members and users of an org can be granted roles on credentials

Only Super Admin can change the private credentials to organizational credentials.

### Job Templates

A job template is a definition and set of parameters for running an Ansible job. Job templates are useful to execute the same job many times.
